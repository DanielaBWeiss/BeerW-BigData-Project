```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
dates = pd.read_csv("../data/calendarific-holidays-2018-_all_.csv")
```

```python
dates = dates[dates["Country"] == "Canada"].reset_index()
```

```python
# Toronto is in Ontario
#all holidays are based on this page https://en.wikipedia.org/wiki/Public_holidays_in_Canada and couple more I have added
```

```python
dates
```

```python
look_into_dates = dates[dates["Name"].isin(["Remembrance Day","All Saints' Day","Halloween","Good Friday","St. Patrick's Day","St David's Day","Valentine's Day","Day After New Year’s Day","New Year's Eve","New Year's Day","Good Friday","Canada Day","Labour Day","Christmas Day","Christmas"])]
look_into_dates
```

```python
holidays = dates[dates["Types"] == "National holiday"]
holidays = holidays.append(look_into_dates)
holidays
```

```python
# march break is not in the dates I found although it is in the wiki page. it is only for school though
#dates[dates["Month"].isin([3])]
```

```python
holidays.drop_duplicates(subset="Name",keep='first', inplace=True) # drop duplicate row
holidays.drop("index",axis=1,inplace=True)
holidays.reset_index(inplace=True,drop=True)
holidays
```

```python
holidays.iloc[:,[0]] = pd.to_datetime(holidays['Date'],infer_datetime_format=True)
```

```python
holidays = holidays.sort_values(by=['Date']) 
holidays.reset_index(inplace=True,drop=True)
holidays
```

```python
holidays_2019 = holidays.copy()
```

```python
holidays_2019["Date"] = holidays_2019["Date"] + pd.to_timedelta(365, unit='D')
holidays_2019.Year += 1
```
